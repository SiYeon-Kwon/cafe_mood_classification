```python
import glob
import cv2
import json
import os
from collections import OrderedDict
```

```python

img = cv2.imread("C:/Users/USER/Desktop/indust/0.jpeg")
height = img.shape[0]
width = img.shape[1]
x = width/2
y = height/2

file_data=OrderedDict()
annotations=OrderedDict()
json_data=OrderedDict()

file_data['image']="66.jpeg"

annotations["label"]="industrial"
annotations["coordinates"]={"x":x,"y":y,"width":width,"height":height}

file_data['annotation']=[annotations]
json_data=[file_data]
print(json.dumps(json_data, ensure_ascii=False, indent="\t") )
```

```python
dir_ = "C:/Windows/System32/kwontest/experiment/train/natural/"

for filename in glob.glob("C:/Windows/System32/kwontest/experiment/train/natural/*.jpeg"):
    basename = os.path.basename(filename)
    print(basename)
    img = cv2.imread(dir_+basename)
    height = img.shape[0]
    width = img.shape[1]
    x = width/2
    y = height/2

    file_data=OrderedDict()
    annotations=OrderedDict()
    json_data=OrderedDict()

    file_data['image']=basename

    annotations["label"]="natural"
    annotations["coordinates"]={"x":x,"y":y,"width":width,"height":height}

    file_data['annotation']=[annotations]
    json_data=[file_data]
    basename=basename.replace("jpeg","json")
    print(basename)

    with open(dir_+basename ,"w") as f:
        json.dump(json_data, f)
```
